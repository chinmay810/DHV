```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={
    "Date":pd.to_datetime(["2022-01-01","2022-02-01","2022-03-01","2022-04-01","2022-05-01","2022-06-01"]),
    "Value":[10,15,22,28,35,42]
      }

df=pd.DataFrame(data)

df.set_index("Date",inplace=True)

plt.figure(figsize=(10,6))
plt.plot(df.index,df["Value"],marker='o')
plt.xlabel("Data")
plt.ylabel("Value")
plt.title("Time series with Trend")
plt.grid(True)
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={
    "Date":pd.to_datetime(["2022-01-01","2022-02-01","2022-03-01","2022-04-01","2022-05-01","2022-06-01","2022-07-01","2022-08-01","2022-09-01","2022-10-01",]),
    "Value":[10,18,15,22,30,42,35,54,42,56]
      }

df=pd.DataFrame(data)

df.set_index("Date",inplace=True)

plt.figure(figsize=(10,6))
plt.plot(df.index,df["Value"],marker=' ')
plt.xlabel("Data")
plt.ylabel("Value")
plt.title("Time series with Trend")
plt.grid(True)
plt.show()
```
